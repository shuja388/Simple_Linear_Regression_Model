```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
#Reading Data...
df = pd.read_csv("Salary_Data.csv")
df.head(5)

#Spliting Data for training & Testing.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['YearsExperience']],df.Salary,test_size=0.3,random_state= 0)

#Fitting Model for traing on Data 70%....
lr = linear_model.LinearRegression().fit(X_train,y_train)

#ploting Training Data..
plt.scatter(X_train,y_train,color='red',marker='+')
plt.xlabel('YearsExperience',color='red',fontsize=16)
plt.ylabel('Salary',color='red',fontsize=16)
plt.title('Training Data',fontsize=20,color='blue')
plt.plot(X_train,lr.predict(X_train),color='blue')
plt.show()

#ploting Testing Data....
plt.scatter(X_test,y_test,color='red')
plt.xlabel('YearsExperience',color='red',fontsize=16)
plt.ylabel('Salary',color='red',fontsize=16)
plt.title('Testing Data',fontsize=20)
plt.plot(X_test,lr.predict(X_test),color='Green')
plt.show()

#Model score evaluating...
print("Traing Data socre :",lr.score(X_train,y_train))
print("Testing Data socre :",lr.score(X_test,y_test))

# Model Prediction on unseen data...
lr.predict([[12],[15],[18],[20]])

#Bar plot showcse....
sns.barplot(y_test)
plt.xlabel('YearsExperience',color='red',fontsize=16)
plt.ylabel('Salary',color='red',fontsize=16)
plt.title('Test Data',fontsize=20,color='blue')
plt.show()
print("Model w(Coef) is:",lr.coef_)
print("Model B(intercept) is:",lr.intercept_)
```
